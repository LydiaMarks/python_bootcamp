Numpy arrays

```python
import numpy as np
```

```python
my_np_array = np.array([[1, 2, 3], [4, 5, 6]], np.int32)
print(type(my_np_array))
# output: <class 'numpy.ndarray'>
print(my_np_array.shape)
# output: (2, 3)
print(my_np_array.dtype)
# output: dtype('int32')
```

notation is always [row, column]
my_np_array = np.array([[row0/column0, row0/column1, row0/column2], [row1/column0, row1/column1, row1/column2]], np.int32)


my_np_array[1, 2] 

This is how we call the element of my_np_array that's in the *second* row, *third* column, namely, 6. Recall that we start at 0, so 1 repesents the 2nd row.




```python
row0_col0 = my_np_array[0,0]
print(row0_col0)

row0_col1 = my_np_array[0,1]
print(row0_col1)

row0_col2 = my_np_array[0,2]
print(row0_col2)

row1_col0 = my_np_array[1,0]
print(row1_col0)

row1_col1 = my_np_array[1,1]
print(row1_col1)

row1_col2 = my_np_array[1,2]
print(row1_col2)

# row2_col1 = my_np_array[2,1]
# print(row2_col1) #this gives an index error because the array is only 2 rows wide (row 0 and row 1) so there is no row 2
```
